```python
# 导入所需要的模块和库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.model_selection import train_test_split, GridSearchCV,RepeatedKFold
from sklearn.metrics import roc_auc_score,roc_curve,auc
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate,KFold
from xgboost import XGBClassifier as XGBC
from sklearn.linear_model import LogisticRegression as LR
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

from time import time
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("集成学习特征筛选.csv")
data
```

```python
X = data.iloc[:,1:-1]
y = data.iloc[:,-1:]
# 切分训练集和测试集
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=23)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
```

```python
clf = XGBC()    #实例化GBDT分类器，并使用默认参数
clf = clf.fit(Xtrain,np.ravel(Ytrain))
train_score = clf.score(Xtrain,Ytrain)
test_score = clf.score(Xtest,Ytest)
print(f"XGBoost在训练集上的预测准确率为{train_score}")
print(f"XGBoost在测试集上的预测准确率为{test_score}")
```

```python
lr = LR(random_state=23) #实例化lr
lr = lr.fit(Xtrain,np.ravel(Ytrain))
score_lr = lr.score(Xtest,Ytest)

svm = SVC(random_state=23) #实例化svm
svm = svm.fit(Xtrain,np.ravel(Ytrain))
score_svm = svm.score(Xtest,Ytest)

rfc = RFC(random_state=23) #实例化随机森林
rfc = rfc.fit(Xtrain,np.ravel(Ytrain))
score_rfc = rfc.score(Xtest,Ytest)

gbc = GBC(random_state=23) #实例化GBDT
gbc = gbc.fit(Xtrain,np.ravel(Ytrain))
score_gbc = gbc.score(Xtest,Ytest)

xgb = XGBC(random_state=23) #实例化XGBC
xgb  = xgb .fit(Xtrain,np.ravel(Ytrain))
score_xgb  = xgb.score(Xtest,Ytest)

print("LR：{}".format(score_lr))
print("SVM：{}".format(score_svm))
print("随机森林：{}".format(score_rfc))
print("GBDT：{}".format(score_gbc))
print("----777----")
print("XGBoost：{}".format(score_xgb))
```

```python
# 画出决策树、随机森林和GBDT在十组五折交叉验证下的效果对比
score_lr = []
score_svm = []
score_xgb = []
score_rfc = []
score_gbc = []

for i in range(10):
    lr = LR()
    cv0 = cross_val_score(lr,X,np.ravel(y),cv=5)
    score_lr.append(cv0.mean())
    
    svm = SVC()
    cv00 = cross_val_score(svm,X,np.ravel(y),cv=5)
    score_svm.append(cv00.mean())
    
    xgb = XGBC()
    cv1 = cross_val_score(xgb,X,np.ravel(y),cv=5)
    score_xgb.append(cv1.mean())
    
    rfc = RFC()
    cv2 = cross_val_score(rfc,X,np.ravel(y),cv=5)
    score_rfc.append(cv2.mean())
    
    gbc = GBC()
    cv3 = cross_val_score(gbc,X,np.ravel(y),cv=5)
    score_gbc.append(cv3.mean())

plt.plot(range(1,11),score_lr,label = "LR")
plt.plot(range(1,11),score_svm,label = "SVM")
plt.plot(range(1,11),score_xgb,label = "XGBoost")
plt.plot(range(1,11),score_rfc,label = "RandomForest")
plt.plot(range(1,11),score_gbc,label = "GBDT")
plt.legend(bbox_to_anchor=(1.4,1))
plt.show()
```

```python
# best = 20
MSE_train = []
MSE_test = []
gamma = np.arange(0,50,5)
overfit = []

for i in gamma:
    xgb = XGBC(gamma=i,random_state=0)
    result = cross_validate(xgb
                            ,X,y,cv=5
                            ,return_train_score=True
                            ,verbose=False) 
    MSE_train.append(abs(result["train_score"].mean()))
    MSE_test.append(abs(result["test_score"].mean()))
    overfit.append(abs(result["test_score"].mean())-abs(result["train_score"].mean()))
                 
plt.plot(gamma,MSE_train,color="green",label="XGB_train")
plt.plot(gamma,MSE_test,color="green",linestyle="--",label="XGB_test")
plt.plot(gamma,overfit,color="orange",label="Overfit")
plt.legend(bbox_to_anchor=(1.4,1))
plt.show()
```

```python
# 绘制lambda学习曲线 best = 20
MSE_train = []
MSE_test = []
lambda_ = np.arange(0,100,10)
overfit = []

for i in lambda_:
    xgb = XGBC(reg_lambda=i,random_state=0)
    result = cross_validate(xgb
                            ,X,y,cv=5
                            ,return_train_score=True
                            ,verbose=False) 
    MSE_train.append(abs(result["train_score"].mean()))
    MSE_test.append(abs(result["test_score"].mean()))
    overfit.append(abs(result["test_score"].mean())-abs(result["train_score"].mean()))
                 
plt.plot(lambda_,MSE_train,color="green",label="XGB_train")
plt.plot(lambda_,MSE_test,color="green",linestyle="--",label="XGB_test")
plt.plot(lambda_,overfit,color="orange",label="Overfit")
plt.legend(bbox_to_anchor=(1.4,1))
plt.show()
```

```python
# 绘制n_estimators的学习曲线 best = 27
27
# ===========【TIME WARNING：2min】============#
scores = []
for i in np.arange(1,50,2):
    xgb = XGBC(n_estimators=i)
    s = cross_val_score(xgb,X,np.ravel(y),cv=5).mean()
    scores.append(s)

print(f"当n_estimators={np.arange(1,50,2)[scores.index(max(scores))]}时，准确率取得最大值{max(scores)}")
plt.plot(np.arange(1,50,2),scores,label = "XGBC")
plt.show()
```

```python
# 绘制不同n_estimators、learning_rate下，模型效果变化
plt.figure(figsize=(10,6))
for i in [27,28,29]:
    scores = []
    for j in np.arange(0.01,0.2,0.01):
        xgb = XGBC(n_estimators=i,learning_rate=j,random_state=23)
        s = cross_val_score(xgb,X,np.ravel(y),cv=5).mean()
        scores.append(s)
    plt.plot(np.arange(0.01,0.2,0.01),scores,label = f"n_estimators={i}")
    plt.legend()
print(max(scores))
```

```python
from sklearn.model_selection import cross_validate,KFold
#定义所需的交叉验证方式
cv = KFold(n_splits=5,shuffle=True,random_state=23)

modelname = ["GBDT","XGBoost","SVM"]
colors = ["green","orange","red"]
clf_models = [GBC(random_state=23),XGBC(random_state=23),SVC(random_state = 23)]

xaxis = range(1,6)
plt.figure(figsize=(10,6),dpi=65)

for name,model,color in zip(modelname,clf_models,colors):
    result = cross_validate(model,X,np.ravel(y),cv=cv
                            ,return_train_score=True)
    plt.plot(xaxis,result["train_score"], color=color, label = name+"_Train")
    plt.plot(xaxis,result["test_score"], color=color, linestyle="--",label = name+"_Test")

plt.xticks([1,2,3,4,5])
plt.xlabel("CVcounts",fontsize=12)
plt.ylabel("Accuracy",fontsize=12)
plt.title("GBDT vs XGBoost vs SVM")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV,RepeatedKFold
# 2次5折交叉验证
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,np.ravel(y),test_size=0.3,random_state=23)
rkf = RepeatedKFold(n_splits=5,n_repeats=2)
area1 = 0 
for train_index,test_index in rkf.split(X):
    Xtrain = X.iloc[train_index]
    Xtest = X.iloc[test_index]
    Ytrain = y.iloc[train_index]
    Ytest = y.iloc[test_index]
    model_xgb = XGBC(n_estimators=27,learning_rate=0.09).fit(Xtrain,Ytrain)
    print("训练集上的预测准确率为：",model_xgb.score(Xtrain,Ytrain))
    print("测试集上的预测准确率为：",model_xgb.score(Xtest,Ytest))
    area = roc_auc_score(Ytest, model_xgb.predict(Xtest))
    print("AUC为：",area)
    area1 += area
    print(area1/10)
    print("---------++++++----------")
    model_xgb_pred = model_xgb.predict(Xtest)
    print(confusion_matrix(Ytest,model_xgb_pred))
    print("----混淆矩阵如上所示----")
    print(classification_report(Ytest,model_xgb_pred))
    FPR, recall, thresholds = roc_curve(Ytest, model_xgb.predict_proba(Xtest)[:,1])
    # 绘制ROC曲线
    plt.figure(dpi = 600)
    plt.plot(FPR, recall, color='red',
            label='ROC curve (area = %0.2f)' % area)
    plt.plot([0, 1], [0, 1], color='black', linestyle='--')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('Recall')
    plt.title('XGBoost ROC curve')
    plt.legend(loc="lower right")
    plt.show()
```
