```python
# 导入所需要的模块和库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
import keras
from keras import Sequential
from keras import layers
from keras import regularizers


from time import time
import warnings
warnings.filterwarnings("ignore")
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("数据清洗.csv")
# 提取特征和标签
X = data.iloc[:,1:-1] 
y = data.iloc[:,-1]
X_row = data.iloc[:,1:-1]
X = X.astype(np.float64)
colName = X.columns
# X = StandardScaler().fit_transform(X)
# X = pd.DataFrame(X)
X.columns = colName
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=0)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
Xtrain = StandardScaler().fit_transform(Xtrain)
Xtest = StandardScaler().fit_transform(Xtest)
Xtrain = pd.DataFrame(Xtrain)
Xtest = pd.DataFrame(Xtest)
```

```python
#使用lassoCV自带的正则化路径长度和路径中的alpha个数来自动建立alpha选择的范围
alphas = np.logspace(0.001,2,300)
model_lassoCV = LassoCV(eps=0.17
              ,n_alphas=1
              ,cv=5).fit(Xtrain,Ytrain)
print(model_lassoCV.alpha_)
# print(model_lassoCV.alphas_)
print(model_lassoCV.score(Xtest,Ytest))
print("-----****-----")
print(model_lassoCV.coef_)
```

```python
coef = pd.Series(model_lassoCV.coef_,index = X.columns)
print(coef)
print("Lasso picked "+str(sum(coef != 0))+" variables and eliminated the other "+str(sum(coef==0)))
index = coef[coef != 0].index
X = X[index]
# print(X.head())
print(coef[coef != 0])
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=0)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
Xtrain = StandardScaler().fit_transform(Xtrain)
Xtest = StandardScaler().fit_transform(Xtest)
Xtrain = pd.DataFrame(Xtrain)
Xtest = pd.DataFrame(Xtest)
```

```python
# DNN/ANN
model = Sequential()
model.add(layers.Dense(2,kernel_regularizer=regularizers.l2(0.05),input_dim = 20,activation="relu"))
model.add(layers.Dense(23,kernel_regularizer=regularizers.l2(0.05),activation="relu"))
model.add(layers.Dense(23,kernel_regularizer=regularizers.l2(0.05),activation="relu"))

model.add(layers.Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["acc"])
# 训练模型
hist = model.fit(Xtrain,Ytrain,epochs = 300,validation_data=(Xtest,Ytest))
```

```python
plt.plot(range(300),hist.history.get("acc"))
plt.plot(range(300),hist.history.get("loss"))
plt.show()
```

```python
loss,acc = model.evaluate(Xtrain,Ytrain)
loss,acc = model.evaluate(Xtest,Ytest)
```

```python
plt.plot(hist.epoch,hist.history.get("acc"))
plt.plot(hist.epoch,hist.history.get("val_acc"))
plt.show()
```

```python
# 减少网络结构
model = Sequential()
model.add(layers.Dense(2,input_dim = 20,activation="relu"))
model.add(layers.Dense(2,activation="relu"))

model.add(layers.Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["acc"])
# 训练模型
hist = model.fit(Xtrain,Ytrain,epochs = 300,validation_data=(Xtest,Ytest))
```

```python
plt.figure(dpi = 123)
plt.plot(hist.epoch,hist.history.get("acc"))
plt.plot(hist.epoch,hist.history.get("val_acc"))
plt.title("减少网络结构")
plt.xlabel("迭代次数")
plt.ylabel("ACC")
plt.show()
```

```python
# dropout
model = Sequential()
model.add(layers.Dense(23,input_dim = 20,activation="relu"))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(23,activation="relu"))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(23,activation="relu"))
model.add(layers.Dropout(0.5))

model.add(layers.Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["acc"])
# 训练模型
hist = model.fit(Xtrain,Ytrain,epochs = 300,validation_data=(Xtest,Ytest))
```

```python
plt.figure(dpi = 123)
plt.plot(hist.epoch,hist.history.get("acc"))
plt.plot(hist.epoch,hist.history.get("val_acc"))
plt.title("Dropout")
plt.xlabel("迭代次数")
plt.ylabel("ACC")
plt.show()
```
