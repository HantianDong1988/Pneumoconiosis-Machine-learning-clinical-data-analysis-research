```python
# 导入相应模块和包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split


plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("数据清洗.csv")
data.info()
```

```python
# 提取特征和标签
X = data.iloc[:,1:-1]
y = data.iloc[:,-1]
```

```python
# 切分训练集和测试集
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
Xtrain
```

```python
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
```

```python
#Lasso进行拟合
lasso_1 = Lasso(alpha=0.01).fit(Xtrain,Ytrain)
#(lasso_1.coef_*100).tolist()
[*zip(lasso_1.get_feature_names(X.columns),lasso_1.coef_)]
```

```python
lasso_2 = Lasso(alpha=0.05).fit(Xtrain,Ytrain)
(lasso_2.coef_*100).tolist()
```
