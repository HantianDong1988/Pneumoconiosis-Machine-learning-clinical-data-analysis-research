```python
# 导入所需要的模块和库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import keras
from keras import Sequential
from keras import layers
from keras import regularizers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


from time import time
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("data/单因素.csv")
# 提取特征和标签
X = data.iloc[:,:-1] 
y = data.iloc[:,-1]
X_row = data.iloc[:,:-1]
X = X.astype(np.float64)
colName = X.columns
# X = StandardScaler().fit_transform(X)
# X = pd.DataFrame(X)
X.columns = colName
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=0)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
Xtrain = StandardScaler().fit_transform(Xtrain)
Xtest = StandardScaler().fit_transform(Xtest)
Xtrain = pd.DataFrame(Xtrain)
Xtest = pd.DataFrame(Xtest)
```

```python
Xtrain.shape,Ytrain.shape
```

```python
Ytrain.value_counts()
```

```python
# DNN/ANN
model = Sequential()
model.add(layers.Dense(23,kernel_regularizer=regularizers.l2(0.05),input_dim = 16,activation="relu"))
# model.add(layers.Dense(23,kernel_regularizer=regularizers.l2(0.05),activation="relu"))
# model.add(layers.Dense(23,kernel_regularizer=regularizers.l2(0.05),activation="relu"))
```

```python
model.add(layers.Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["acc"])
```

```python
# 训练模型
hist = model.fit(Xtrain,Ytrain,epochs = 300,validation_data=(Xtest,Ytest))
```

```python
plt.plot(range(300),hist.history.get("acc"))
```

```python
plt.plot(range(300),hist.history.get("loss"))
```

```python
loss,acc = model.evaluate(Xtrain,Ytrain)
loss,acc = model.evaluate(Xtest,Ytest)
```

```python
plt.plot(hist.epoch,hist.history.get("acc"),color = "r")
plt.plot(hist.epoch,hist.history.get("val_acc"),color = "g")
plt.show()
```
