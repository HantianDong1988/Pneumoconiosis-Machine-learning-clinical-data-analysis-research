```python
# 导入相应模块和包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split, GridSearchCV,RepeatedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error # 均方误差
from scipy.stats import pearsonr,ttest_ind,levene
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import classification_report

plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv("数据清洗.csv")
```

```python
X_row = data.iloc[:,1:-1]
data_cf = data[data["是否尘肺"]==1]
data_jc = data[data["是否尘肺"]==0]
```

```python
# t检验特征筛选
counts = 0
index = []
for colName in data.columns[1:-1]:
    if levene(data_cf[colName],data_jc[colName])[1]>0.05:
        if ttest_ind(data_cf[colName],data_jc[colName])[1]<0.05:
            counts +=1
            index.append(colName)
        else:
            if ttest_ind(data_cf[colName],data_jc[colName],equal_var=False)[1]<0.05:
                counts +=1
                index.append(colName)
print(len(colName));
print(index)
```

```python
from random import shuffle
if "是否尘肺" not in index:index = ["是否尘肺"]+index
data_cf = data_cf[index]
data_jc = data_jc[index]
data = pd.concat([data_cf,data_jc])
# 提取特征和标签
X = data.iloc[:,1:] 
y = data.iloc[:,0]
X = X.astype(np.float64)
colName = X.columns
X = StandardScaler().fit_transform(X)
X = pd.DataFrame(X)
X.columns = colName
```

```python
alphas = np.logspace(-3,2,100)
model_lassoCV = LassoCV(alphas=alphas,cv = 5,max_iter=100000).fit(X,y)
print(model_lassoCV.alpha_)
```

```python
MSEs = (model_lassoCV.mse_path_)

MSEs_mean,MSEs_std = [],[]
for i in range(len(MSEs)):
    MSEs_mean.append(MSEs[i].mean())
    MSEs_std.append(MSEs[i].std())

#MSEs_mean = np.apply_along_axis(np.mean,1,MSEs)
#MSEs_std = np.apply_along_axis(np.std,1,MSEs)

plt.figure(dpi = 600)
plt.errorbar(model_lassoCV.alphas_,MSEs_mean # x,y 一一对应
            ,yerr=MSEs_std  # y误差范围
            ,fmt="o"    #数据点标记
            ,ms = 2     # 数据点大小
            ,mfc = "r"  # 数据点颜色
            ,mec = "r"  # 数据点边缘颜色
            ,ecolor = "lightblue"  # 误差棒颜色
            ,elinewidth =1     # 误差棒线宽
            ,capsize = 4     # 误差棒边界线长度
            ,capthick = 1)    # 误差棒边界线厚度
plt.semilogx()
plt.axvline(model_lassoCV.alpha_,color = "black",ls="--")
plt.xlabel("Lambda")
plt.ylabel("MSE")
# ax = plt.gca()
# y_major_locator =MultipleLocator(0.5)
# ax.yaxis.set_major_locator(y_major_locator)
plt.show()
```

```python
coef = pd.Series(model_lassoCV.coef_,index = X.columns)
print(coef)
print("Lasso picked "+str(sum(coef != 0))+" variables and eliminated the other "+str(sum(coef==0)))
```

```python
index = coef[coef != 0].index
X = X[index]
# print(X.head())
print(coef[coef != 0])
```

```python
coefs = model_lassoCV.path(X_row,y,alphas=alphas,max_iter = 100000)[1].T
plt.figure(dpi = 600)
plt.semilogx(model_lassoCV.alphas_,coefs,"-")
plt.axvline(model_lassoCV.alpha_,color = "black",ls = "--")
plt.xlabel("Lambda")
plt.ylabel("Coefficients")
plt.show()
```

```python
from sklearn.linear_model import Lasso
coefs = []
for alpha in alphas:
    model_lasso = Lasso(alpha = alpha,max_iter = 100000).fit(X_row,y)
    coefs.append(model_lasso.coef_)
plt.figure(dpi = 600)
plt.semilogx(alphas,coefs,"-")
plt.axvline(model_lassoCV.alpha_,color = "black",ls = "--")
plt.xlabel("Lambda")
plt.ylabel("Coefficients")
plt.show();
```

```python
#特征筛选后做图
x_values = np.arange(len(index))
y_values = coef[coef != 0]
plt.figure(figsize=(16,8),dpi = 600)
plt.bar(x_values,y_values
       ,color = "lightblue"
       ,edgecolor = "black"
       ,alpha = 0.8
       )
plt.xticks(x_values,index
          ,rotation="90"
          ,ha = "right"
          ,va = "top"
          )
plt.xlabel("feature")
plt.ylabel("weight")
plt.show()
```

```python
# 导出数据
X_y = pd.concat([X,y],axis = 1)
X_y.to_csv("集成学习特征筛选.CSV")
```

```python
# 随机森林
# 切分训练集和测试集
from sklearn.metrics import roc_auc_score,roc_curve,auc
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=23)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
clf = RFC() #实例化随机森林
clf_f = clf.fit(Xtrain,np.ravel(Ytrain))
print("训练集上的预测准确率为：",clf_f.score(Xtrain,Ytrain))
print("测试集上的预测准确率为：",clf_f.score(Xtest,Ytest))
area = roc_auc_score(Ytest,clf_f.predict(Xtest))
print("AUC为：",area)
```

```python
clf_pred = clf_f.predict(Xtrain)
print(confusion_matrix(Ytrain,clf_pred))
print("----混淆矩阵如上所示----")
print(classification_report(Ytrain,clf_pred))
```

```python
FPR, recall, thresholds = roc_curve(Ytest,clf_f.predict_proba(Xtest)[:,1])
# 绘制ROC曲线
plt.figure(dpi = 600)
plt.plot(FPR, recall, color='red',
         label='ROC curve (area = %0.2f)' % area)
plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('Recall')
plt.title('RF ROC curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#支持向量机
from sklearn.svm import SVC
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=23)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
model_svm = SVC(kernel="rbf",gamma="auto",probability=True).fit(Xtrain,Ytrain)
print("训练集上的预测准确率为：",model_svm.score(Xtrain,Ytrain))
print("测试集上的预测准确率为：",model_svm.score(Xtest,Ytest))
area = roc_auc_score(Ytest,model_svm.predict(Xtest))
print("AUC为：",area)
```

```python
# 支持向量机网格搜索
Cs = np.logspace(-1,3,10,base = 2)
gammas = np.logspace(-4,1,50,base = 2)
param_grid = dict(C = Cs,gamma = gammas)
grid = GridSearchCV(SVC(kernel="rbf"),param_grid=param_grid,cv=5).fit(Xtrain,Ytrain)
print(grid.best_params_)
C = grid.best_params_["C"]
gamma = grid.best_params_["gamma"]
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.3,random_state=23)
for i in [Xtrain,Xtest,Ytrain,Ytest]:
    i.reset_index(drop = True,inplace = True)
model_svm = SVC(kernel="rbf",C=C,gamma=gamma,probability=True).fit(Xtrain,Ytrain)
print("训练集上的预测准确率为：",model_svm.score(Xtrain,Ytrain))
print("测试集上的预测准确率为：",model_svm.score(Xtest,Ytest))
area = roc_auc_score(Ytest,model_svm.predict(Xtest))
print("AUC为：",area)
```

```python
model_svm_pred = model_svm.predict(Xtrain)
print(confusion_matrix(Ytrain,model_svm_pred))
print("----混淆矩阵如上所示----")
print(classification_report(Ytrain,model_svm_pred))
```
